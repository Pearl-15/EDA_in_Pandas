### 7.2.Explore Correlation

```python
df.info()
```

```python
# unable to direct use .corr() cuz there are object data type which causes pandas unable to calcuate correlation, 
# explicity need to tell pandas to skip object data type when calculate correlation
# df.corr()  => this will cause error

# correct way
df.select_dtypes(include=['number']).corr()
```

```python
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True)

plt.rcParams[ 'figure.figsize'] = (20,7)

plt.show()
```

### 7.3.Find out which continent is growing fastest

```python
# group by continent

df.groupby('Continent').max()
```

```python
df.columns
```

```python
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values( by = '2022 Population', ascending=False)
df2
```

```python
# switch x and y => Continent as column and 2022 Population , 2020 Population as row/index
df3 = df2.transpose()
df3
```

```python
# visualize
df3.plot()
```

### 7.4.Find out top 5 countries that are growing fast

```python
df4 = df.groupby(['Country'])[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values( by='1970 Population', ascending=False).head()
df4
```

```python
df5 = df4.transpose()
df5
```

```python
df5.plot()
```
